```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import display
from frame_sample import VideoFrameSampler  # Replace with the actual module name or paste the class into the notebook
```

```python

#Step 1: Enter your local video path here
video_path = '../../data/test.mp4'  # e.g., 'my_video.mp4'

# Step 2: Instantiate the sampler
sampler = VideoFrameSampler(video_path)

# Step 3: Run sampling methods
uniform_frames = sampler.uniform_sampling(interval=60)
random_frames = sampler.random_sampling(num_frames=5)
sequential_frames = sampler.sequential_sampling(start_frame=100, end_frame=200, step=10)
scene_frames = sampler.scene_based_sampling(threshold=100.0)

sampler.release()

# Step 4: Helper to show a few frames from each method
def show_frames(frames, title, max_frames=5):
    plt.figure(figsize=(15, 3))
    for i, frame in enumerate(frames[:max_frames]):
        plt.subplot(1, min(len(frames), max_frames), i + 1)
        # Convert BGR to RGB for display
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(rgb_frame)
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

# Step 5: Display frames from each method
show_frames(uniform_frames, "Uniform Sampling")
show_frames(random_frames, "Random Sampling")
show_frames(sequential_frames, "Sequential Sampling")
show_frames(scene_frames, "Scene-Based Sampling")
```
